```python
from langchain.agents.agent_types import AgentType
from langchain.chat_models import ChatCohere
from langchain_experimental.agents.agent_toolkits import create_pandas_dataframe_agent
import pandas as pd
from langchain.llms import Cohere
import phoenix as px
from phoenix.trace.langchain import LangChainInstrumentor, OpenInferenceTracer
```

```python
session = px.launch_app()
```

```python
# Instrument the LangChain agent so LLM tracing can start in the background
LangChainInstrumentor().instrument()
```

```python
# Add the file location
arxiv_data = pd.read_csv("../arxiv100.csv")
```

```python
arxiv_data['label'].unique()
```

```python
arxiv_data.info()
```

```python
# Create pandas agent from LangChain library
agent = create_pandas_dataframe_agent(Cohere(cohere_api_key = 'enter-your-Cohere-API-key',temperature = 0), arxiv_data, verbose=True)
```

Start running the queries you want to ask your dataset as a agent

```python
agent.run('What is this dataframe about? Tell me the basic statistics')
```

```python
agent.run('What is dark matter?')
```

```python
agent.run('What do black holes look like?')
```
